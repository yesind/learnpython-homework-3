```python
from collections import Counter 
import pandas as pd

stations = pd.read_csv(
    open('bus_stops.csv', 'r', encoding='cp1251'),
    sep=';'
)

list_stations = stations["Name"]
char1 = '», '
char2 = ' ('
res_list=[]
for station in list_stations:
    res_list.append(station[station.find(char1)+3 : station.find(char2)])
streets = Counter(res_list)
sorted(streets.items(), key=lambda item: item[1], reverse=True)
```
